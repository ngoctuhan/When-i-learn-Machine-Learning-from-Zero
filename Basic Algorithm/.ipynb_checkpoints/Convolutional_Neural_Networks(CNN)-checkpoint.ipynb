## Convolutional Neural Networks là gì?


### Convolutional là gì?

Là một cửa sổ trượt (Sliding Windows) trên một ma trận như mô tả hình dưới:


![](images/Convolutional.png)

Các convolution layer có các tham số (parameter(kernel)) đã được học lấy ra nhưng thông tin chính xác nhất mà không cần chọn các đặc trưng (feature). Sliding Window hay còn gọi là kernel, filter hoặc feature detect là một ma trận có kích thước nhỏ như trong ví dụ trên là 3x3. Convolution hay tích chập là nhân từng phần tử bên trong ma trận 3x3 với ma trận bên trái. Kết quả được một ma trận gọi là Convoled feature được sinh ra từ việc nhận ma trận Filter với ma trận ảnh 5x5 bên trái.

### Convolutional Neral Network là gì?

CNN được tạo thành từ các tế bào thần kinh (neurons) và có khả năng học tập các tham số weight và biases. Mỗi một tế bào thần kinh nhận được một số đầu vào và mỗi tế bào thần kinh thực hiện chấm điểm cho các đầu vào đó, thao tác này thực hiện với hàm phi tuyến tính.  Ví dụ như một mạng sẽ biểu diễn điểm số của một pixel hình ảnh thô với các lớp và nó cũng có một hàm mất mát tương tự như SVM hay Softmax.

Vậy sự khác biệt giữa ConvNet là các mạng thần kinh thông thường là gì?

Kiến trúc ConvNet đưa ra giả định rõ ràng rằng các đầu vào là hình ảnh, cho phép chúng ta mã hóa các thuộc tính nhất định vào kiến trúc. Những điều này làm cho hàm số của các tế bào thần kinh sau đó thực tiện hiệu quả hơn và giảm đáng kể số lượng tham số trong mạng.



## Tổng quan kiến thức.

Regular Neural Nets: Là mạng thần kinh có đầu vào là một vector và biến đổi vector đó thông qua các hidden layer. Mỗi hidden layer có các tế bào thần kinh(neurons) các tế bào thần kinh này có đầu vào là đầu ra của các tế bào thần kinh của lớp trước ( hay nói một cách khác toàn bộ các neurons đều được nối với các neurons của lớp trước nó). Trong cùng một hidden layer thì các neurons hoạt động một cách độc lập. Lớp kết nối cuối cùng được gọi là output layer. Và trong bài toán phân loại nó đặc trưng cho điểm số của mỗi lớp.

Ý tưởng của ConvNets:

Một tập dữ liệu đầu vào ta cần tìm được các feature đặc trưng của các lớp. Sau đó với mỗi hình ảnh mới được đưa vào ta dùng một cửa sổ trượt trên hình ảnh đầu vào đó, và kiểm tra với các feature đặc trưng. Điều này coi như hiệu quả khi hình ảnh chúng ta bị nghiêng hay là nằm lệch về các góc.

CNNs so sánh hình ảnh theo từng mảnh. Các mảnh mà nó tìm được gọi là các feature. Bằng cách tìm ở mức thô các feature khớp nhau ở cùng vị trí trong hai hình ảnh, CNNs nhìn ra sự tương đồng tốt hơn nhiều so với việc khớp toàn bộ bức ảnh.

## Một số khái niệm trong CNNs
### Padding

Khi chúng ta trượt cửa số lần lượt trên hình ảnh, các pixel ở rìa, cạnh của ảnh được sử dụng với số lần ít nhất có thể là 1 lần, như vậy các thông tin của ảnh tại các khu vực này có thể bị thiếu xót vì vậy chúng ta cần sử dụng đệm trước khi tổng hợp sản phẩm.

Đệm được sử dụng bằng cách thêm các pixel vào các cạnh ( thường 1 pixel). Chẳng hạn hình ảnh kích thước 6 * 6 chúng ta sau thêm lớp đệm nó sẽ có kích thước là 8 * 8

### Strided Convolutions

Stride chính là sải chân của của sổ trượt từ trái sang phải và từ trên xuống dưới trên toàn bộ bức ảnh.

Khi xem một hình ảnh mới, CNN không biết chính xác nơi các feature này sẽ khớp nên nó sẽ thử chúng khắp mọi nơi, ở mọi vị trí có thể. Khi tính toán sự khớp của một feature trên toàn bộ ảnh, chúng ta làm thành một filter (bộ lọc). Phần toán ta sử dụng để làm điều này được gọi là tích chập, từ đó mà Mạng Nơ-ron Tích chập (Convolutional Neural Networks) có tên là như vậy.

Phép toán được sử dụng đây là phép nhân từng điểm ảnh tương ứng với nhau. 

+ RELU sẽ áp dụng hàm kích hoạt theo phần tử :

![](images/relu.png)

Đóng vai trò nhỏ nhưng quan trọng trong quá trình này là Rectified Linear Unit hoặc ReLU. Toán của cái này cũng rất đơn giản bất cứ nơi nào có số âm, hoán đổi nó với 0. Điều này giúp CNN giữ vững sự tin cậy toán học bằng cách giữ các giá trị đã được học khỏi bị mắc kẹt gần 0 hoặc bị thổi bay về vô tận.

+ POOL (để chung) :

![](images/poo.png)

Pooling là một cách lấy những hình ảnh lớn và làm co chúng lại trong khi vẫn giữ các thông tin quan trọng nhất trong đó.Nó bao gồm việc duyệt bước một ô vuông cửa sổ nhỏ dọc trên một hình ảnh và lấy giá trị lớn nhất từ cửa sổ ở mỗi bước. Trong thực tế, một cửa sổ có cạnh 2 hoặc 3 điểm ảnh và duyệt bước mỗi 2 điểm ảnh là được.

+ Lớp FC (tức là được kết nối đầy đủ) sẽ tính toán điểm số của lớp, dẫn đến khối lượng kích thước [1x1x10]

#### Lớp Convolutional

Đây là một lớp cốt lỗi xây dựng một Neural Networks. 
Tại lớp Conv thì các bộ lọc (kenrnel) có kích thước nhỏ sẽ trượt qua input đầu vào và tính tổng tích chập của filter và slide windows trên imput, chúng đưa ra một điểm số dưới dạng một ma trận 2 chiều biểu diễn điểm chấm của bộ lọc tại bất kì vị trí nào.
Nếu ai đó hỏi việc tìm các filter này như thế nào? Câu hỏi là tôi chịu, việc đó sẽ có thư viện hỗ trợ chúng ta việc tìm kiếm và tính toán để đưa ra các filter.
Các node trong neutrol network xuất hiện ở đâu trọng quá trình này. Các node đó chính là điểm số của các vị trị được tính toán dựa trên filter. Các node cũng có thể được hiểu là một noron chỉ nhìn vào một phần của đầu vào và chia sẻ tham số với các noron khác bên trái và bên phải của nó.

Sự xắp xếp không gian của các noron :

Có một sự khá là phức tạp ở đây tôi sẽ cố gắng làm nó dễ hiểu.

![](images/depthcol.jpeg)




Trên hình có 5 noron đang nhìn vào một khu vực đầu vào nhưng đầy đủ với độ sâu đầu vào. Như vậy tại mỗi khu vực tau đều sẽ có các cột nhìn vào một khu vực trên đầu vào input. Mô hình này giống với bánh kem xốp chúng ta hay ăn chúng cho nhiều tâng. Bây giờ chúng ta đục các lỗ trên bề mặt của chiếc bánh, khi tách từng lớp ra ta vẫn thấy các lỗ trên các lớp, nó tương tự như chấm điểm input bằng filter , các lớp  sau là các lớp thu nhỏ đặc trưng của lớp trước tạo nên cái điển hình nhất.



Nhưng có bạn nào tự hỏi ra là làm thế nào để không gian đầu ra bằng với không gian đầu vào sau mỗi phép Conv. Câu trả lời sẽ có sau đây. Đó là chúng ta thêm các padding (đệm vào input) chúng ta bao quanh input bằng các đệm. Tại sao vậy tau phải biết thêm một thứ đó là stride (sải chân) có nghĩa là mỗi lần slide windows trượt qua bao nhiêu pixel được quyết định ở tham số này. 
Một đầu vào có kích thước 5*5 tôi sẽ đêm xung quanh nó các giá trị 0 khi đó tôi sẽ được đầu vào với kích thước là 6*6, stride = 1, kernel = 3 * 3 tôi sẽ được một ma trận điểm số cho bộ filter tại mọi vị trí có kích thước là 5 * 5 nếu không tin bạn hãy vẽ bàn cờ và nhìn chúng. Có thể tham khảo hình sau đây:

![](iamges/stride.gif)

Kích thước đầu ra của một lớp Conv được xác định như sau:
$(W - F + 2P)/S + 1$

Trong đó :
W là kích thước của ma trận đầu vào.
F là kích thước bộ lọc
S là độ lớn của dải chân
P là số lượng phần đềm sử dụng.

Ví dụ: đối với đầu vào 7x7 và bộ lọc 3x3 với bước 1 và pad 0, chúng ta sẽ nhận được đầu ra  (7 - 3 + 2 * 0)/1 + 1 = 5x5. Với sải chân 2 chúng ta sẽ có được đầu ra 3x3.

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle 
import os
from scipy.misc import imread
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        datadict = pickle.load(fo, encoding='bytes')
    
    Y = datadict[b'labels']
    X = datadict[b'data']
    
    X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("float")
    Y = np.array(Y)
    return X, Y
def load_CIFAR10(ROOT):
    xs = []
    ys = []
    for b in range(1,6):
        f = os.path.join(ROOT, 'data_batch_%d' % (b, ))
        X, Y = unpickle(f)
        xs.append(X)
        ys.append(Y)    
    Xtr = np.concatenate(xs)
    Ytr = np.concatenate(ys)
    del X, Y
    Xte, Yte = unpickle(os.path.join(ROOT, 'test_batch'))
    return Xtr, Ytr, Xte, Yte
```

```python
cifar10_dir = 'F:\DataML\cifar-10-python\cifar-10-batches-py'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

# As a sanity check, we print out the size of the training and test data.
print ('Training data shape: ', X_train.shape)
print ('Training labels shape: ', y_train.shape)
print ('Test data shape: ', X_test.shape)
print ('Test labels shape: ', y_test.shape)
```

```python
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
samples_per_class = 7
for y, cls in enumerate(classes):
    idxs = np.flatnonzero(y_train == y)
    idxs = np.random.choice(idxs, samples_per_class, replace=False)
    for i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx)
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
plt.show()
```

```python
def conv2D(x,W,b,stride = 1):
    x = tf.nn.conv2d(x, W, strides = [1, stride,stride, 1], padding = "SAME")
    x = tf.nn.bias_add(x, b)
    return tf.nn.relu(x)
def maxpool(x, k = 2):
    return tf.nn.max_pool(x, ksize=[1, k, k, 1], strides=[1, k, k, 1],padding='SAME')
```

Lớp tích chập đầu tiên có các bộ lọc 32-3x3, thì khóa đầu tiên ( wc1) trong từ điển trọng số có một đối số shapecó một giá trị với 4 giá trị: đầu tiên và là kích thước bộ lọc, trong khi thứ ba là số lượng kênh trong hình ảnh đầu vào và cuối cùng đại diện cho số lượng bộ lọc tích chập bạn muốn trong lớp chập đầu tiên. Khóa đầu tiên trong biasestừ điển bc1, sẽ có 32 tham số sai lệch.
Tương tự cho 2 lớp tích chập

Sau khi áp dụng 3 thao tác tích chập và gộp tối đa, bạn đang ghép xuống hình ảnh đầu vào từ 32 x 32 x 3 đến 4 x 4 x 128 (stride = 2) và bây giờ bạn cần phải làm phẳng đầu ra được ghép xuống này để cung cấp điều này làm đầu vào cho lớp được kết nối đầy đủ. Đó là lý do tại sao bạn thực hiện thao tác nhân 4 * 4 * 128 là đầu ra của lớp trước hoặc số kênh được xuất ra bởi lớp chập 3. Phần tử thứ hai của bộ dữ liệu mà bạn truyền vào shape có số lượng tế bào thần kinh mà bạn muốn trong lớp kết nối đầy đủ. Tương tự, trong biasestừ điển, khóa thứ tư bd1 có 128 tham số.

```python
weights = {
    'wc1': tf.get_variable('W0', shape=(3,3,1,32), initializer=tf.contrib.layers.xavier_initializer()), 
    'wc2': tf.get_variable('W1', shape=(3,3,32,64), initializer=tf.contrib.layers.xavier_initializer()), 
    'wc3': tf.get_variable('W2', shape=(3,3,64,128), initializer=tf.contrib.layers.xavier_initializer()), 
    'wd1': tf.get_variable('W3', shape=(4*4*128,128), initializer=tf.contrib.layers.xavier_initializer()), 
    'out': tf.get_variable('W6', shape=(128,n_classes), initializer=tf.contrib.layers.xavier_initializer()), 
}
biases = {
    'bc1': tf.get_variable('B0', shape=(32), initializer=tf.contrib.layers.xavier_initializer()),
    'bc2': tf.get_variable('B1', shape=(64), initializer=tf.contrib.layers.xavier_initializer()),
    'bc3': tf.get_variable('B2', shape=(128), initializer=tf.contrib.layers.xavier_initializer()),
    'bd1': tf.get_variable('B3', shape=(128), initializer=tf.contrib.layers.xavier_initializer()),
    'out': tf.get_variable('B4', shape=(10), initializer=tf.contrib.layers.xavier_initializer()),
}
```

Bây giờ chúng ta băt đầu ghép nối các thần phần của CNNs lại thành một mạng hoàn chỉnh 

```python
def CNNs(x, weights, biases):
    
    # Convolution layer 
    conv1 = conv2D(x, weights['wc1'], biases['bc1'], stride=2)
    conv1 = maxpool(conv1, k = 2)
    
    conv2 = conv2D(conv1, weights['wc2'], biases['bc2'], stride=2)
    conv2 = maxpool(conv2, k = 2)
    
    conv3 = conv2D(conv2, weights['wc3'], biases['bc3'], stride=2)
    conv3 = maxpool(conv3, k = 2)
    
    # full conection
    fc1 = tf.reshape(conv3, [-1, weights['wd1'].get_shape().as_list()[0]])
    
    fc1 = tf.add(tf.matmul(fc1, weights['wd1']), biases['bd1'])
    
    fc1 = tf.nn.relu(fc1)
    
    out = tf.add(tf.matmul(fc2, weights['out']) , biases['out'])
    
    return out

```
